```python
import pandas as pd 
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv(r'C:\Users\dimri\OneDrive\Desktop\vs code java\myfirstvscode\logistic regression\robosheet2.csv')
```

```python
df = df.drop(['Evaporation', 'Sunshine','Cloud9am','Cloud3pm','Location','Date'], axis = 1)
#we drop all these columns 
df.head()
```

```python
df = df.dropna(axis =0)
df.shape
```

```python
categorical_columns = ['WindGustDir', 'WindDir9am', 'WindDir3pm', 'RainToday', 'RainTomorrow']

for col in categorical_columns:
    df[col] = pd.factorize(df[col])[0]
```

```python
df.head()
```

```python
df.shape
```

```python
df.head()
```

```python
km = KMeans(n_clusters = 10)
df['cluster'] = km.fit_predict(df)
```

```python
df.head()
```

```python
df.cluster.value_counts()
```

this is not a realistic approach to determine how many clusters there should be. so we use elbow method

```python
dist_df=[]
for num_clusters in range(1,10):
    km = KMeans(n_clusters=num_clusters)
    km.fit(df)
    dist_df.append([num_clusters,km.inertia_])
dist_df = pd.DataFrame(dist_df,columns=['number_of_clusters','distance'])
```

```python
dist_df
```

```python
dist_df.set_index('number_of_clusters').plot()
plt.xlabel('Number of clusters')
plt.ylabel('distance (squared) to cluster center')
plt.title('K_Means elbow method')
```
